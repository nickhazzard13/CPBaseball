```python
engine = get_engine()
# https://pandas.pydata.org/docs/reference/api/pandas.read_sql.html
# Note: since our column names are capitalized, you must wrap each column in quotes
df1 = pd.read_sql("""
    SELECT *
    FROM all_college_2025
    WHERE "PitcherTeam" = 'CAL_POL';
""", con=engine)

df1
```

```python
df = pd.read_sql("""
    SELECT "PitcherTeam", COUNT(*) AS n
    FROM all_college_2025
    GROUP BY "HitterTeam"
    ORDER BY n DESC;
""", con=engine)

df
```
