```python
import pandas as pd
from sqlalchemy import text

# --- STEP 1: Load team data into a clean DataFrame ---
TEAM = "CAL_MUS"  # change if needed

# If you already defined get_engine() earlier, this will just work.
engine = get_engine()

df_raw = pd.read_sql(
    text('SELECT * FROM all_college_2025 WHERE "BatterTeam" = :team'),
    con=engine,
    params={"team": TEAM},
)

print("Rows loaded:", len(df_raw))
print("Columns:", len(df_raw.columns))
print("Sample columns:", list(df_raw.columns)[:12])

# Quick peek at unique batters with batted-ball data
num_cols = ["ExitSpeed","Angle","Direction","Distance","PositionAt110X","PositionAt110Y"]
for c in num_cols:
    if c in df_raw.columns:
        df_raw[c] = pd.to_numeric(df_raw[c], errors="coerce")

batters = (
    df_raw.loc[df_raw["ExitSpeed"].notna(), "Batter"]
          .dropna()
          .drop_duplicates()
          .sort_values()
          .tolist()
)

print("Batters with batted-ball data (first 10):", batters[:10])
print("Total batters with batted-ball data:", len(batters))

# Small preview
display(df_raw.head(5))
```

```python
# --- STEP 2: Compute X/Y field coordinates and sanity-check with a plot ---

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image


# Optional: path to your field image (PNG recommended). Use None to skip.
FIELD_IMAGE = "/Users/nicholashazzard/CPBaseball/baseball.png"  # e.g., "/Users/you/CPBaseball/baseball.png"

# Work on a copy
df = df_raw.copy()

# Ensure numeric types for key columns
for c in ["ExitSpeed","Angle","Direction","Distance","PositionAt110X","PositionAt110Y"]:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# Keep only batted balls
hits = df[df["ExitSpeed"].notna()].copy()

# Primary XY from Direction (deg) + Distance (ft)
# Convention assumed: 0° = straightaway CF, negative = pull (LF), positive = oppo (RF)
use_dir_dist = hits["Distance"].notna() & hits["Direction"].notna()
theta = np.deg2rad(hits.loc[use_dir_dist, "Direction"])
D = hits.loc[use_dir_dist, "Distance"]
hits.loc[use_dir_dist, "X_ft"] = D * np.sin(theta)
hits.loc[use_dir_dist, "Y_ft"] = D * np.cos(theta)

# Fallback XY from PositionAt110X/Y (TrackMan’s 110-ft plane)
fallback = hits["X_ft"].isna() & hits["PositionAt110X"].notna() & hits["PositionAt110Y"].notna()
hits.loc[fallback, "X_ft"] = hits.loc[fallback, "PositionAt110X"]
hits.loc[fallback, "Y_ft"] = hits.loc[fallback, "PositionAt110Y"]

# Drop rows still missing XY
hits_xy = hits.dropna(subset=["X_ft","Y_ft"]).copy()

print(f"Usable batted balls with XY: {len(hits_xy)} / {len(hits)}")

# ---- Quick sanity plot ----
X_MIN, X_MAX = -330, 330
Y_MIN, Y_MAX = 0, 420

fig, ax = plt.subplots(figsize=(7,7))

if FIELD_IMAGE:
    try:
        img = Image.open(FIELD_IMAGE).convert("RGBA")
        img = img.transpose(Image.FLIP_TOP_BOTTOM)   # 👈 flip vertically so home plate is at the bottom
        ax.imshow(np.array(img), extent=[X_MIN, X_MAX, Y_MIN, Y_MAX], origin="lower", zorder=0)
    except Exception as e:
        print(f"Could not load field image: {e}")


sc = ax.scatter(
    hits_xy["X_ft"], hits_xy["Y_ft"],
    c=hits_xy["ExitSpeed"], s=20, alpha=0.9, cmap="plasma", zorder=1
)
ax.set_xlim(X_MIN, X_MAX); ax.set_ylim(Y_MIN, Y_MAX)
ax.set_aspect("equal","box")
ax.set_xlabel("Horizontal (ft)  [− = LF, + = RF]")
ax.set_ylabel("Depth (ft)  [0 = Home, + = CF]")
ax.set_title("Spray Chart — All Batters (sanity check)")
cbar = plt.colorbar(sc, ax=ax, fraction=0.046, pad=0.04); cbar.set_label("Exit Velocity (mph)")
plt.show()

# Peek at a few batters to ensure values look reasonable
print("Example batters:", hits_xy["Batter"].dropna().unique()[:5])
```

```python
# --- STEP 3: Single-batter sanity check + KPIs and charts ---

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image

# Choose a batter and a min EV threshold
batters = sorted(hits_xy["Batter"].dropna().unique().tolist())
print("Batters found (first 10):", batters[:10])
BATTER = batters[0] if batters else None   # <- set to a specific name if you want
MIN_EV = 50

# Optional: field image path
FIELD_IMAGE = "/Users/nicholashazzard/CPBaseball/baseball.png"

# Filter batter + threshold
sub = hits_xy[(hits_xy["Batter"] == BATTER) & (hits_xy["ExitSpeed"] >= MIN_EV)].copy()
print(f"{BATTER}: {len(sub)} BIP (EV ≥ {MIN_EV}) after filters")

# ---- KPIs ----
def fmt(n, unit=""):
    return "–" if pd.isna(n) else (f"{n:.1f}{unit}" if isinstance(n, (float, int)) else str(n))

ev_avg = sub["ExitSpeed"].mean()
ev_max = sub["ExitSpeed"].max()
la_avg = sub["Angle"].mean() if "Angle" in sub and sub["Angle"].notna().any() else np.nan
hh_rate = (sub["ExitSpeed"] >= 95).mean() if len(sub) else np.nan

print("KPIs")
print("BIP:", len(sub))
print("Avg EV:", fmt(ev_avg, " mph"))
print("Max EV:", fmt(ev_max, " mph"))
print("Avg LA:", fmt(la_avg, "°"))
print("Hard-Hit %:", "–" if pd.isna(hh_rate) else f"{100*hh_rate:.0f}%")

# ---- A) Polar spray (Direction vs Distance) ----
req = sub[["Direction","Distance","ExitSpeed"]].dropna()
fig = plt.figure(figsize=(6.5,6.5))
ax = fig.add_subplot(111, polar=True)
if len(req):
    theta = np.deg2rad(req["Direction"].to_numpy())
    r = req["Distance"].to_numpy()
    sc = ax.scatter(theta, r, c=req["ExitSpeed"], s=36, alpha=0.9)
    ax.set_theta_zero_location('N'); ax.set_theta_direction(-1)
    ax.set_title(f"Polar Spray — {BATTER} (EV ≥ {MIN_EV})")
    cbar = plt.colorbar(sc, ax=ax, pad=0.1); cbar.set_label("Exit Velo (mph)")
else:
    ax.set_title("Polar Spray — no Direction/Distance available")
plt.show()

# ---- B) Field overlay (X_ft, Y_ft on image) ----
X_MIN, X_MAX = -330, 330
Y_MIN, Y_MAX = 0, 420

fig, ax = plt.subplots(figsize=(7,7))
# draw field image if available
try:
    img = Image.open(FIELD_IMAGE).convert("RGBA")
    img = img.transpose(Image.FLIP_TOP_BOTTOM)  # keep home at bottom
    ax.imshow(np.array(img), extent=[X_MIN, X_MAX, Y_MIN, Y_MAX], origin="lower", zorder=0)
except Exception as e:
    print(f"Field image not loaded: {e}")

plotdf = sub.dropna(subset=["X_ft","Y_ft"])
if len(plotdf):
    sc = ax.scatter(plotdf["X_ft"], plotdf["Y_ft"], c=plotdf["ExitSpeed"], s=36, alpha=0.9, cmap="plasma", zorder=1)
    cbar = plt.colorbar(sc, ax=ax, fraction=0.046, pad=0.04); cbar.set_label("Exit Velo (mph)")
else:
    ax.text(0.5,0.5,"No usable XY for this selection", ha="center", va="center", transform=ax.transAxes)

ax.set_xlim(X_MIN, X_MAX); ax.set_ylim(Y_MIN, Y_MAX)
ax.set_aspect("equal","box")
ax.set_xlabel("Horizontal (ft)  [− = LF, + = RF]")
ax.set_ylabel("Depth (ft)  [0 = Home, + = CF]")
ax.set_title(f"Field Overlay — {BATTER} (EV ≥ {MIN_EV})")
plt.show()
```

```python
# --- STEP 4: Generate spray charts for multiple hitters ---

import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import pandas as pd

# Choose up to 10 batters
batters = sorted(hits_xy["Batter"].dropna().unique().tolist())[:10]
print("Plotting these batters:", batters)

FIELD_IMAGE = "/Users/nicholashazzard/CPBaseball/baseball.png"

# Load and flip field image once
img = Image.open(FIELD_IMAGE).convert("RGBA")
img = img.transpose(Image.FLIP_TOP_BOTTOM)
field_array = np.array(img)

X_MIN, X_MAX = -330, 330
Y_MIN, Y_MAX = 0, 420

# Create grid of subplots (adjust layout if fewer than 10)
n = len(batters)
cols = 5
rows = int(np.ceil(n / cols))

fig, axes = plt.subplots(rows, cols, figsize=(cols * 3.5, rows * 3.5))
axes = axes.flatten()

for i, batter in enumerate(batters):
    sub = hits_xy[(hits_xy["Batter"] == batter) & (hits_xy["ExitSpeed"] >= 50)].dropna(subset=["X_ft","Y_ft"])
    ax = axes[i]
    ax.imshow(field_array, extent=[X_MIN, X_MAX, Y_MIN, Y_MAX], origin="lower", zorder=0)
    ax.set_xlim(X_MIN, X_MAX)
    ax.set_ylim(Y_MIN, Y_MAX)
    ax.set_aspect("equal", "box")
    ax.set_title(batter, fontsize=8)
    ax.axis("off")

    if len(sub):
        sc = ax.scatter(sub["X_ft"], sub["Y_ft"], c=sub["ExitSpeed"], s=10, alpha=0.9, cmap="plasma", zorder=1)

# Remove any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

fig.suptitle("Spray Charts — Top 10 Batters (EV ≥ 50 mph)", fontsize=14, y=0.95)
fig.tight_layout()
plt.show()
```

```python
# --- Pitcher Movement Profile using AutoPitchType ---

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import text

TEAM = "CAL_MUS"  # change if needed
engine = get_engine()

# 1) Load team pitching data
cols_needed = [
    "Pitcher","PitcherThrows","PitcherTeam","AutoPitchType",
    "InducedVertBreak","HorzBreak","RelSpeed"
]
q = text(f'''
    SELECT {", ".join(f'"{c}"' for c in cols_needed)}
    FROM all_college_2025
    WHERE "PitcherTeam" = :team
''')
dfp = pd.read_sql(q, engine, params={"team": TEAM})

# 2) Clean + keep rows with movement, velo, and auto pitch type
for c in ["InducedVertBreak","HorzBreak","RelSpeed"]:
    dfp[c] = pd.to_numeric(dfp[c], errors="coerce")

dfp["AutoPitchType"] = dfp["AutoPitchType"].astype(str).str.strip()

valid_types = {"Four-Seam","Sinker","Cutter","Changeup","Splitter","Slider","Curveball","Other"}
dfp = dfp[
    dfp["AutoPitchType"].isin(valid_types) &
    dfp["InducedVertBreak"].notna() &
    dfp["HorzBreak"].notna() &
    dfp["RelSpeed"].notna() &
    dfp["Pitcher"].notna()
].copy()

# 3) Pick a pitcher
pitchers = sorted(dfp["Pitcher"].unique().tolist())
print("Pitchers (first 10):", pitchers[:10])
PITCHER = pitchers[0] if pitchers else None  # set a specific name if you want

sub = dfp[dfp["Pitcher"] == PITCHER].copy()
if sub.empty:
    raise ValueError("No pitches for selected pitcher.")

throws = (sub["PitcherThrows"].dropna().iloc[0] if sub["PitcherThrows"].notna().any() else "R").upper()

# (Optional) normalize handedness: make +HB = arm side like a RHP
NORMALIZE_HANDEDNESS = True
if NORMALIZE_HANDEDNESS and throws.startswith("L"):
    sub["HorzBreak"] = -sub["HorzBreak"]

# 4) Summaries: usage % and velo by pitch
summary = (
    sub.groupby("AutoPitchType")
       .agg(Count=("AutoPitchType","size"), MPH=("RelSpeed","mean"))
       .sort_values("Count", ascending=False)
       .reset_index()
)
summary["Usage %"] = (100 * summary["Count"] / summary["Count"].sum()).round(0).astype(int)
summary["MPH"] = summary["MPH"].round(1)

# 5) Colors per AutoPitchType (tweak as you like)
COLORS = {
    "Four-Seam": "#d9534f",
    "Sinker":    "#f0ad4e",
    "Cutter":    "#b294bb",
    "Changeup":  "#88d498",
    "Splitter":  "#8dd3c7",
    "Slider":    "#f0e442",
    "Curveball": "#5bc0de",
    "Other":     "#7f8c8d",
}

# 6) Plot: HB (x) vs IVB (y)
def plot_movement_profile(df_pitcher, name, throws_hand):
    # axis limits
    max_abs = np.nanmax(np.abs(df_pitcher[["HorzBreak","InducedVertBreak"]].to_numpy()))
    lim = int(np.ceil(max(24, max_abs + 4) / 6.0) * 6)   # to nearest 6"
    lim = min(lim, 36)
    ticks = [-24, -12, 0, 12, 24]

    fig, ax = plt.subplots(figsize=(8,8))

    # background rings (12" and 24")
    for r in (12, 24):
        ax.add_patch(plt.Circle((0,0), r, color="#cfd8dc", fill=False, lw=1.5, alpha=0.7))

    # crosshairs
    ax.axhline(0, color="#b0bec5", lw=1.2)
    ax.axvline(0, color="#b0bec5", lw=1.2)

    # clusters
    for pt, g in df_pitcher.groupby("AutoPitchType"):
        c = COLORS.get(pt, "#7f8c8d")
        ax.scatter(g["HorzBreak"], g["InducedVertBreak"], s=35, c=c, alpha=0.9, label=pt, edgecolors="none")

    ax.set_xlim(-lim, lim); ax.set_ylim(-lim, lim)
    ax.set_aspect("equal","box")
    ax.set_xticks(ticks); ax.set_yticks(ticks)
    ax.set_xlabel('Horizontal Break (")  [− = glove-side, + = arm-side]')
    ax.set_ylabel('Induced Vertical Break (")  [↑ more rise, ↓ more drop]')
    ax.set_title(f"{pd.Timestamp.today().year} Movement Profile (Induced Break)\n{name} — Throws: {throws_hand}", pad=12)

    ax.legend(title="Pitch", loc="upper right", frameon=False, fontsize=9)
    plt.tight_layout()
    return fig, ax

fig, ax = plot_movement_profile(sub, PITCHER, throws)
plt.show()

# 7) Show summary table
display(summary.style.hide(axis="index"))
```
